```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
 
image=cv2.imread("yoda.jpeg")
m = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(m,cmap='gray')
```
